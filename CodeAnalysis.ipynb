Load data from files generated by http://www.spinellis.gr/sw/ckjm/ (Coupling metrics) and
http://cyvis.sourceforge.net/ (Cycllomatic Complexity) and merge them into one file

```python
metricsFile = "out\\metrics.csv"

merger = CBMerger("data\\exampleCA.csv", "data\\exampleComplexity.csv", metricsFile)
merger.merge_metrics()
```

Initiate code base analyzer with the merged data

```python
analyzer = CBAnalyzer(metricsFile)
```

```python
%matplotlib notebook
analyzer.orig_scatter_plot()
```

Now we have inverted metrics. Here is the plot of inverted and normalized values of Coupling and Complexity:

$$
NormComplexity = \frac{1+min(Complexity)}{1+Complexity}∗100
$$

$$
NormCA = \frac{1+min(CA)}{1+CA}∗100
$$


```python
%matplotlib notebook
analyzer.inverted_scatter_plot()
```

Last step is to scalarize our criteria and save everything into the output file.
Here is the formula for the criteria:

$$
Rank(NormComplexity , NormCA)=100−\frac{\sqrt{NormComplexity^2+NormCA^2}}{\sqrt{2}}
$$
